```python
# Load env variables
from dotenv import load_dotenv, get_key, dotenv_values

load_dotenv()
```

```python
# Create and configure an anthropic client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Make a request
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "O que é computação quântica? Responda em apenas uma frase."
        }
    ]    
)
```
