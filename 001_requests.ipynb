```python
# Load env variables
from dotenv import load_dotenv, get_key, dotenv_values

load_dotenv()
```

```python
# Create and configure an anthropic client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Make a request
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "O que é computação quântica? Responda em apenas uma frase."
        }
    ]    
)
```

```python
message.content[0].text
```

```python
# Claude doesn't store context of conversation
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "write another sentence"
        }
    ]    
)
message.content[0].text
```

```python
def add_user_message(messages: list, text: str):
    user_message={"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages: list, text: str):
    assistant_message={"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages: list):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return message.content[0].text
```

```python
# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "Defina computação quântica em uma frase")

# Get Claude's response
answer = chat(messages)

# Add Claude's response to the conversation history
add_assistant_message(messages, answer)

# Add a follow-up question
add_user_message(messages, "Escreva outra frase")
 
# Get the follow-up response with full context
final_answer = chat(messages)
```

```python
# Add the second response from assintant to history
add_assistant_message(messages, final_answer)
```

```python
# Chat bot exercise

messages = []

while True:
    user_input=input("> ")
    print(">", user_input)

    add_user_message(messages, user_input)
    answer=chat(messages)
    add_assistant_message(messages,answer)
    print(f"---\n{answer}\n---\n")
```
