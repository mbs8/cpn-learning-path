```python
# Imports, client configuration and helper functions

from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()
model = "claude-sonnet-4-6"

def add_user_message(messages: list, text: str):
    user_message={"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages: list, text: str):
    assistant_message={"role": "assistant", "content": text}
    messages.append(assistant_message)

# now this method has a parameter system to configure the prompts of the assistant
def chat(messages: list, system=None):
    params={
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }

    if(system != None):
        params["system"]=system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# without system prompts
messages=[]

add_user_message(messages, "How do I solve 5x+3=2 for x?")
answer=chat(messages)
print(answer)
```

```python
# with system prompts
messages=[]

system_prompt = """
You are a patient math tutor.
Do not directly answer a student's questions.
Guide them to a solution step by step.
"""

add_user_message(messages, "How do I solve 5x+3=2 for x?")

# adding system prompt to chat method
answer=chat(messages, system=system_prompt)
print(answer)
```

```python
# System prompt exercise

# first try to write a python function without any prompt
messages=[]

add_user_message(
    messages,
    "Write a python function that checks a string for duplicate characters.",
)

answer=chat(messages)
print(answer)
```

```python
# now using the system prompt to reduce and optimize code
messages=[]

add_user_message(
    messages,
    "Write a Python function that checks a string for duplicate characters.",
)

system_prompt="""
You are a Python developer expert.
Always write the most concise code with good documentation.
Do not add comments in the middle of code.
Preserve the good stucture and follow the best coding patterns.
Emphasis on code performance.
"""

answer=chat(messages, system=system_prompt)
print(answer)
```
