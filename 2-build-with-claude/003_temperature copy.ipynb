```python
# Imports, client configuration and helper functions

from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()
model = "claude-sonnet-4-6"

def add_user_message(messages: list, text: str):
    user_message={"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages: list, text: str):
    assistant_message={"role": "assistant", "content": text}
    messages.append(assistant_message)

# now this method also has a temperature param to configure randomness/creativity of claude responses
# from 0.0 to 1.0, where 0 is deterministic and 1.0 is more random/creative
def chat(messages: list, system=None, temperature=1.0):
    params={
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
    }

    if(system != None):
        params["system"]=system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# using temperature to adjust claude creativity
messages=[]

add_user_message(
    messages,
    "Create a videogame idea in one sentence",
)

answer=chat(messages, temperature=0.0)
print(answer)
```

```python
answer=chat(messages, temperature=0.0)
print(answer)
```

```python
answer=chat(messages, temperature=1.0)
print(answer)
```

```python
answer=chat(messages, temperature=1.0)
print(answer)
```
